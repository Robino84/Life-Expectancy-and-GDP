```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data 

df = pd.read_csv("all_data.csv")

#show the first 10 rows 
print(df.head(10))

#print data types 
print(df.info())

#Summary stats 
print(df.describe()) 
```

```python
# Rename the Life Expectancy column 
df.rename({"Life expectancy at birth (years)": "Life_Expectancy"}, axis="columns", inplace=True)

#Create a new column (GDP in USD billions) 
df["GDP_Billions"] = df["GDP"] / 1000000000

#Show the first 5 rows of the new dataframe 
print(df.head())
```

```python
# Has life expectancy increased over time in the six nations?

#Plot for all countries 

ax = sns.lineplot(x = "Year", y = "Life_Expectancy", hue = "Country", style = "Country", data = df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in different countries")
plt.show()
plt.clf()

#Plot for Chile 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Chile"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Chile")
plt.show()
plt.clf()

#Plot for China 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "China"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in China")
plt.show()
plt.clf()

#Plot for Germany
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Germany"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Germany")
plt.show()
plt.clf()

#Plot for Mexico
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Mexico"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Mexico")
plt.show()
plt.clf()

#Plot for USA 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "United States of America"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in USA")
plt.show()
plt.clf()

#Plot for Zimbabwe 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Zimbabwe"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Zimbabwe")
plt.show()
plt.clf()
```

In the plots above, we can see that life expectancy decreased in Zimbabwe between 2001 and 2004. Then it started to increase between 2005 and 2015. However, in the latest observation (2015), there is still a significant difference in Life Expectancy compared to the other countries in the study. 

In the other countries, Life Expectancy has increased over the time period 2000-2015. Nevertheless, Life Expectancy has not increased every year in all countries. Only in China Life Expectancy increased every single year in the sample period.

```python
#Calculate the correlation matrix for all variables and countries 

correlation = df.corr(method="pearson")

print(correlation)

print(" ")

#Calculate the correlation coefficient for individual countries 

list = ["Chile", "China", "Germany", "Mexico", "United States of America", "Zimbabwe"]

def corr_ind(list):
    for country in list:
        correlation = data = df[df.Country == country].GDP_Billions.corr(df[df.Country == country].Life_Expectancy)
        print("The correlation coefficient between GDP and Life Expectancy in " +country +" is equal to " +str(round(correlation, 2)))

corr_ind(list)   
```

The correlation between GDP and Life Expectancy for all countries is 0.34 (weak association). This is significantly lower compared to the country-specific correlation coefficients that are all higher than 0.9 (very strong association). One possible explanation is that we only have 16 observations per country, and when the sample increases, we get a more stable/reliable correlation. 

If we look at the scatterplot of GDP vs Life Expectancy for China further above, it can be seen that there is an association between GDP and Life Expectancy. Higher GDP values are associated with higher Life Expectancy. However, the association is not linear. 

If we look at the scatterplot of GDP vs Life Expectancy for Zimbabwe further above, the association is not that clear. In some cases, we see that higher GDP values are associated with lower Life Expectancy. 

```python
# What is the average life expectancy in these nations?

def life_ind(list):
    for country in list:
        avg_life = df[df.Country == country]["Life_Expectancy"].mean()
        print("The average life expectancy in " +country +" for the sample period is " +str(round(avg_life, 0)))

life_ind(list)   
```

The average life expectancy for all countries except Zimbabwe lies between 74 and 80 years. For Zimbawe it is significantly lower, only 50 yrs.  

```python
# What is the distribution of life expectancy?

#Boxplot with the variable Life Expectancy 

ax = sns.boxplot(data = df, y = "Life_Expectancy", x="Year") 
ax.set(title = "Distribution of Life Expectancy across all countries")
plt.xticks(rotation= 45)
plt.show()
plt.clf() 

#Boxplot with the variable Life Expectancy, changing the start- and end values on the y-axis in order to see the quartiles more clearly. 
ax = sns.boxplot(data = df, y = "Life_Expectancy", x="Year") 
ax.set(title = "Distribution of Life Expectancy across all countries - Zoomed in on the quartiles")
plt.xticks(rotation= 45)
ax.set_ylim(ymin=70, ymax=82)
plt.show()
plt.clf() 


#Calculating the Interquartile Range


q3 = df.groupby(['Year'])['Life_Expectancy'].apply(lambda x: np.percentile(x,75))
q1 = df.groupby(['Year'])['Life_Expectancy'].apply(lambda x: np.percentile(x,25))
iqr = q3 - q1
print("The IQR for each year:")
print(iqr) 
```

The outliers in the chart with the title "Distribution of Life Expectancy across all countries", represent values of Zimbabwe. 

During the sample period, the lower and upper quartile, as well as the median, climb to higher levels. Moreover, the interquartile range (IQR) is around 4 years for each box plot in the chart.
