```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data 

df = pd.read_csv("all_data.csv")

#show the first 10 rows 
print(df.head(10))

#print data types 
print(df.info())

#Summary stats 
print(df.describe()) 
```

```python
# Rename the Life Expectancy column 
df.rename({"Life expectancy at birth (years)": "Life_Expectancy"}, axis="columns", inplace=True)

#Create a new column (GDP in USD billions) 
df["GDP_Billions"] = df["GDP"] / 1000000000

#Show the first 5 rows of the new dataframe 
print(df.head())
```

```python
# Has life expectancy increased over time in the six nations?

#Plot for all countries 

ax = sns.lineplot(x = "Year", y = "Life_Expectancy", hue = "Country", style = "Country", data = df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in different countries")
plt.show()
plt.clf()

#Plot for Chile 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Chile"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Chile")
plt.show()
plt.clf()

#Plot for China 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "China"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in China")
plt.show()
plt.clf()

#Plot for Germany
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Germany"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Germany")
plt.show()
plt.clf()

#Plot for Mexico
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Mexico"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Mexico")
plt.show()
plt.clf()

#Plot for USA 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "United States of America"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in USA")
plt.show()
plt.clf()

#Plot for Zimbabwe 
ax = sns.lineplot(x = "Year", y = "Life_Expectancy", data = df[df.Country == "Zimbabwe"], marker="*")
ax.set(xlabel = "Year", ylabel = "Life expectancy", title = "Life expectancy over time in Zimbabwe")
plt.show()
plt.clf()
```

In the plots above, we can see that life expectancy decreased in Zimbabwe between 2001 and 2004. Then it started to increase between 2005 and 2015. However, in the latest observation (2015), there is still a significant difference in Life Expectancy compared to the other countries in the study. 

In the other countries, Life Expectancy has increased over the time period 2000-2015. Nevertheless, Life Expectancy has not increased every year in all countries. Only in China Life Expectancy increased every single year in the sample period.   

```python
# Has GDP increased over time in the six nations?

#Plot for all countries 

ax = sns.lineplot(x = "Year", y = "GDP_Billions", hue = "Country", style = "Country", data = df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in different countries")
plt.show()
plt.clf()

#Plot for Chile 
ax = sns.lineplot(x = "Year", y = "GDP_Billions", data = df[df.Country == "Chile"], marker="*")
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in Chile")
plt.show()
plt.clf()

#Plot for China 
ax = sns.lineplot(x = "Year", y = "GDP_Billions", data = df[df.Country == "China"], marker="*")
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in China")
plt.show()
plt.clf()

#Plot for Germany
ax = sns.lineplot(x = "Year", y = "GDP_Billions", data = df[df.Country == "Germany"], marker="*")
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in Germany")
plt.show()
plt.clf()

#Plot for Mexico
ax = sns.lineplot(x = "Year", y = "GDP_Billions", data = df[df.Country == "Mexico"], marker="*")
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in Mexico")
plt.show()
plt.clf()

#Plot for USA 
ax = sns.lineplot(x = "Year", y = "GDP_Billions", data = df[df.Country == "United States of America"], marker="*")
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in USA")
plt.show()
plt.clf()

#Plot for Zimbabwe 
ax = sns.lineplot(x = "Year", y = "GDP_Billions", data = df[df.Country == "Zimbabwe"], marker="*")
ax.set(xlabel = "Year", ylabel = "GDP Billions of USD", title = "GDP over time in Zimbabwe")
plt.show()
plt.clf()
```

Comments: 

1) GDP has increased in all countries over the time period 2000-2015. In other words, GDP is higher in 2015 compared to the year 2000. 

2) Zimbabwe experienced a falling trend line between 2000 and 2008. After 2008, GDP started to increase every year. According to Investopedia, Zimbabwe has suffered from inflation for many years, which has devastated the economy. The country suffered from hyperinflation between 2007 and 2009. In 2009, Zimbabwe abandoned its currency, and started to use currencies from other countries instead. Maybe this action helped and stopped the falling trend in GDP?  

3) China has had a positive GDP growth rate for ever year in the sample period. 

4) The global financial crisis obviously had a negtive impact on the GDP in USA, Mexico, Germany, and Chile.
